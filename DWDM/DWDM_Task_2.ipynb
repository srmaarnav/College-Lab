### Apriori Algorithm:
### $ Notation -->  C_{k} = Candidate \;itemset \;of \;size \; k \; and\; L_{k} =  frequent \;itemset\; of\; size\; k $
### $  C_{k+1} =\; candidates \;generated \;from \;L_{k}$

### $ L_{k +1} = candidates \;in \; C_{k+1} \;satisfying\; minsup $

1. Read the transaction database and get support for each itemset, compare the support with minimum support to generate frequent itemset at level 1.
2.  Use join to generate a set of candidate k-item sets of length $ K+1 ( C_{k+1} ) $at next level.
3. Generate frequent item sets sets of length $ K+1 (L_{k +1})$at next level using minimum support.
     In this step
        a. scan original database  to count support for k+1 candidates
        b.  prune candidates below minsup
4.  Repeat step 2 and 3 until no frequent item sets can be generated.
5. Generate rules form frequent itemsets from level 2 onwards using minimum confidence.

## Goal of Association Rule Mining and its Application to any buisness

### When you apply Association Rule Mining on a given set of transactions T your goal will be to find all rules with:

### 1. Support greater than or equal to min_support
### 2. Confidence greater than or equal to min_confidence

### One of the algorithm for Association Rule Mining implemented here is APRIORI


### Support - defines the popularity of an item

### Confidence - it tells the possibility of how often items X and Y occur together, given the number of times X has occurred

### Strong rules  - If a rule A=>B [Support, Confidence] satisfies min_sup and min_confidence then it is a strong rule.

###  Lift gives the correlation between A and B in the rule A=>B. Correlation shows how one item-set A effects the item-set B.
 ### Lift(A=>B)=   ( Support ) / (Supp(A)Supp(B))


### If the lift is > 1, then A and B are dependent on each other, and the degree of which is given by lift value.

###  Lift indicates certainty of a rule. How much sale of X has increased when Y is sold?

### Lift(X=>Y) = Confidence(X, Y)/Support(Y)
## Example: X => Y [support = 60%, confidence = 90%]
## 60% of all the transactions show that X and Y have been bought together.

## 90% of the customers who bought X also bought Y.

```python
import numpy as np
import pandas as pd
from apyori import apriori

data_frame = pd.read_csv('market_basket.csv', header =None)
data_frame.head()

lsts = []
for i in range (0,22):
    lsts.append([str(data_frame.values[i,j]) for j in range (0,6)])

asscsn_rules = apriori(lsts, min_support =0.50, min_confidence = 0.7, min_lift = 1.2, min_length = 2)
asscsn_results = list(asscsn_rules)

# print(asscsn_results)
formatted_rules = []
for result in asscsn_results:
    for ordered_stat in result.ordered_statistics:
        formatted_rules.append({
            'Base': ', '.join(list(ordered_stat.items_base)),
            'Add': ', '.join(list(ordered_stat.items_add)),
            'Support': result.support,
            'Confidence': ordered_stat.confidence,
            'Lift': ordered_stat.lift
        })

# Convert to a DataFrame for better readability
rules_df = pd.DataFrame(formatted_rules)
print(rules_df)
```

```python
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder

data = pd.read_csv('market_basket.csv', header=None)
transactions = []
for i in range(len(data)):
    transactions.append([str(data.values[i, j]) for j in range(data.shape[1]) if str(data.values[i, j]) != 'nan'])

num_transactions = data.shape[0]

# Use TransactionEncoder to convert the list of transactions into a format suitable for the FP-Growth algorithm.
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

frequent_items = fpgrowth(df, min_support=0.5, use_colnames=True)
print(frequent_items)

rules = association_rules(frequent_items, metric="confidence", min_threshold=0.7, num_itemsets=22)
rules = rules[rules['lift'] >= 1.2]
rules = rules[rules['antecedents'].apply(lambda x: len(x)) + rules['consequents'].apply(lambda x: len(x)) >= 2]
# Reset the index and start numbering from 1
rules.reset_index(drop=True, inplace=True)
rules.index += 1  # Shift the index to start from 1

print(rules)
```
