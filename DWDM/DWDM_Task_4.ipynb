```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Create a DataFrame for easy plotting
df = pd.DataFrame(X, columns=iris.feature_names)

# Scatter plot of sepal length vs sepal width
plt.figure(figsize=(8, 6))
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=y, cmap='viridis', edgecolors='k', s=100)
plt.title('Iris Dataset: Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.colorbar(label='Species')
plt.show()
```

# Write a program to illustrate the k-means clustering algorithm

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load Iris dataset
iris = load_iris()
X = iris.data

# K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Plotting
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, alpha=0.5)
plt.title('K-Means Clustering')
plt.show()
```

# Write a program to illustrate the mini-batch k-means clustering algorithm

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import MiniBatchKMeans

iris = load_iris()
X = iris.data

# Mini-Batch K-Means clustering
minibatch_kmeans = MiniBatchKMeans(n_clusters=3, random_state=42)
y_minibatch_kmeans = minibatch_kmeans.fit_predict(X)

# Plotting
plt.scatter(X[:, 0], X[:, 1], c=y_minibatch_kmeans, s=50, cmap='viridis')
plt.scatter(minibatch_kmeans.cluster_centers_[:, 0], minibatch_kmeans.cluster_centers_[:, 1], c='red', s=200, alpha=0.5)
plt.title('Mini-Batch K-Means Clustering')
plt.show()
```

# Write a program to illustrate the divisive clustering algorithm

```python
from sklearn.cluster import AgglomerativeClustering

# Divisive clustering simulation (splitting larger clusters)
def divisive_clustering(X, max_clusters=3):
    n = len(X)
    current_clusters = [X]

    while len(current_clusters) < max_clusters:
        # Split the largest cluster using AgglomerativeClustering
        largest_cluster = max(current_clusters, key=len)
        clustering = AgglomerativeClustering(n_clusters=2)
        labels = clustering.fit_predict(largest_cluster)

        # Split the cluster into two
        cluster_1 = largest_cluster[labels == 0]
        cluster_2 = largest_cluster[labels == 1]

        current_clusters.remove(largest_cluster)
        current_clusters.extend([cluster_1, cluster_2])

    return current_clusters

# Divisive clustering (splitting the Iris dataset into 3 clusters)
divisive_clusters = divisive_clustering(X, max_clusters=3)

# Plotting divisive clustering
plt.scatter(X[:, 0], X[:, 1], c='blue', s=50, cmap='viridis')
for cluster in divisive_clusters:
    plt.scatter(cluster[:, 0], cluster[:, 1], s=50, cmap='viridis')
plt.title('Divisive Clustering Simulation')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plot
import numpy as np

x = [[i] for i in [12, 10, 6, 8, 4, 2, 1, 16]]

clusters = [x]
while len(clusters) < 3:
    largest_cluster = max(clusters, key=len)
    cluster1 = largest_cluster[:len(largest_cluster)//2]
    cluster2 = largest_cluster[len(largest_cluster)//2:]

    clusters.remove(largest_cluster)
    clusters.extend([cluster1, cluster2])

Z = linkage(x, 'ward')
figure = plot.figure(figsize=(25,10))
den = dendrogram(Z)
plot.show()
```

# Write a program to illustrate the DB-scan clustering algorithm

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data

# DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

# Plotting
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, s=50, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.show()
```
